```python
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook as tqdm

#set the max columns to none"
pd.set_option('display.max_columns', None)
```

```python
size = 'demo'
type_ = 'train'

behaviors_path = f'./files/parquet/ebnerd_{size}/{type_}/behaviors.parquet'
behaviors_train_df = pd.read_parquet(behaviors_path)
print(f'Raw {size} validation behaviors df shape:          ',behaviors_train_df.shape)


size = 'demo'
type_ = 'validation'

behaviors_path = f'./files/parquet/ebnerd_{size}/{type_}/behaviors.parquet'
behaviors_val_df = pd.read_parquet(behaviors_path)
print(f'Raw {size} validation behaviors df shape:          ',behaviors_val_df.shape)
```

```python
behaviors_train_df
```

```python
final_df =  pd.concat([behaviors_train_df, behaviors_val_df], ignore_index=True)[['article_ids_clicked']].explode('article_ids_clicked').value_counts().reset_index()
```

```python
# Save the interaction matrix to pickle
top_ten_file_path = f'./files/pickle/top_10_articles.pkl'

final_df.to_pickle(top_ten_file_path)
```
